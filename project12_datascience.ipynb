```python
import pandas as pd

data = pd.read_csv('./data.csv') # https://www.kaggle.com/karangadiya/fifa19
data
```

```python
#grabbing the value, wage, and player category
df = pd.DataFrame(data, columns=['Name', 'Wage', 'Value'])
df
```

```python
# adding a difference column
def value_to_float(x):
    if type(x) == float or type(x) == int:
        return x
    if 'K' in x:
        if len(x) > 1:
            return float(x.replace('K', '')) * 1000
        return 1000.0
    if 'M' in x:
        if len(x) > 1:
            return float(x.replace('M', '')) * 1000000
        return 1000000.0
    if 'B' in x:
        return float(x.replace('B', '')) * 1000000000
    return 0.0

wage = df['Wage'].replace('[\€,]', '', regex=True).apply(value_to_float) #check regex101
value = df['Value'].replace('[\€,]', '', regex=True).apply(value_to_float)

df['Wage'] = wage
df['Value'] = value

df['Difference'] = df['Value'] - df['Wage']
df
```

```python
df.sort_values('Difference', ascending=False)
```

```python
# visualize
import seaborn as sns
sns.set()

graph = sns.scatterplot(x='Wage', y='Value', data=df)
graph
```

```python
# interactive visualization
from bokeh.plotting import figure, show
from bokeh.models import HoverTool

TOOLTIPS = HoverTool(tooltips=[
    ("Name", "@Name"),
    ('Index', "$index"),
    ("(Wage, Value)", "(@Wage, @Value)")
    ]) # https://docs.bokeh.org/en/latest/docs/user_guide/tools.html?highlight=hovertool

p = figure(title='Soccer 2019',  x_axis_label='Wage', y_axis_label='Value', plot_width=700, plot_height=700, tools=[TOOLTIPS])
p.circle('Wage', 'Value', size=10, source=df)

show(p)
```
